```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
safe_urls=pd.read_csv("C:/Users/HP/Desktop/legitimate-urls.csv")
```

```python
phishing_urls=pd.read_csv("C:/Users/HP/Desktop/phishing-urls.csv")
```

```python
urls=safe_urls.append(phishing_urls)
#merge urls into one dataframe
```

```python
urls.shape[1] #columns
```

```python
urls.shape[0] #rows
```

```python
urls.columns
```

```python
urls=urls.drop(urls.columns[[0,3,5]],axis=1)
```

```python
urls=urls.sample(frac=1).reset_index(drop=True)
urls
#sample(frac=1)- shuffle data randomly for equal distribution
#frac=1 ratio of determines what fraction of total instances need to be returned.
#reset_index(drop=True) - If you set drop = True, the current index will be deleted entirely because it got been shuffled randomly after sample(frac=1) func 
#and the numeric index will replace it.
```

```python
urls_without_labels=urls.drop('label',axis=1)
#dropping labels and storing in dataframe for training_data and test_data
urls_without_labels.columns
```

```python
labels=urls['label']
#storing class labels for training and testing
labels
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, labels_train, labels_test = train_test_split(urls_without_labels,labels,test_size=0.3,random_state=110)
#data_train = 70, labels_train=70, data_test=30, labels_test=30
```

```python
print(len(data_train),len(data_test),len(labels_train),len(labels_test))
#count(data_train)=count(labels_train)
#count(data_test)=count(labels_test)
```

```python
labels_train.value_counts()
#value_counts() function returns object containing counts of unique values. 
#The resulting object will be in descending order so that the first element is the most frequently-occurring element.
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_classifier=RandomForestClassifier()
#generating model
```

```python
random_forest_classifier.fit(data_train,labels_train)
#fit func to rrain our model
```

```python
prediction_label=random_forest_classifier.predict(data_test)
```
